```python
datos= loadtxt('dimer_observations.dat',delimiter=' ')

x = datos[:,0]
y = datos[:,1]
```

```python
def func_prime(y,alpha,bet,Km1,Km2,So):
    
    return (alpha*(So-y)/(Km1+So-y))-(bet*y/(Km2+y))
```

```python
h=0.1
n_points = int((30)/h)
tiempo = []
for i in range(n_points):
    tiempo=append(tiempo,(i+1)*h)
```

```python
def RungeKutta(y0,h,alpha,bet,Km1,Km2,So):
    k1 = func_prime(y0,alpha,bet,Km1,Km2,So)
    
    #first step
    y1 = y0 + (h/2.0) * k1
    k2 = func_prime(y1,alpha,bet,Km1,Km2,So)
    
    #second step
    y2 = y0 + (h/2.0) * k2
    k3 = func_prime( y2,alpha,bet,Km1,Km2,So)
        
    #third step
    y3 = y0 + h * k3
    k4 = func_prime(y3,alpha,bet,Km1,Km2,So)
    
    #fourth step
    average_k = (1.0/6.0)*(k1 + 2.0*k2 + 2.0*k3 + k4)

    y= y0 + h * average_k
    
    return y
```

```python
def Chi2(ysol,y):
    
    chi = sum((ysol-y)**2)
    
    return chi
```
